# Heart Disease Data Visualisation just Coding

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("/content/heart-disease.csv")
heart_disease.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10, 10),
                               sharex = True,
                               frameon = True)


first = ax0.scatter(x = heart_disease["age"],
                    y = heart_disease["sex"],
                    c = heart_disease["target"],
                    cmap = "winter")

ax0.legend(*first.legend_elements(), title = "Target")
ax0.set(title = "Heart Disease Risk Comparison Between Male and Female",
        ylabel = "Sex")

ax0.set_ylim([-0.5, 1.5])

for spines in ax0.spines.values():
  spines.set_edgecolor("brown")

ax0.axhline(y = heart_disease["sex"].mean(), linestyle = "--")

second = ax1.scatter(x = heart_disease["age"],
                     y = heart_disease["oldpeak"],
                     c = heart_disease["target"],
                     cmap = "winter")

ax1.set(title = "Heart Disease and Old Peaks",
        xlabel = "Age",
        ylabel = "Old Peaks")

ax1.legend(*second.legend_elements(), title = "Target")

ax1.set_xlim([30, 80])

ax1.set_ylim([-1, 7])

for spine in ax1.spines.values():
  spine.set_edgecolor("brown")

ax1.axhline(y = heart_disease["oldpeak"].mean(), linestyle = "--")

fig.suptitle("Heart Disease Anaylsis", fontsize = 20, fontweight = "bold", color = "Brown")

fig.patch.set_facecolor("yellow");
```
